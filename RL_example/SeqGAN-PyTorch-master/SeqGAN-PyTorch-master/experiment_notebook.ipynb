```python
import pickle as pkl
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
with open('experiment.pkl', 'rb') as f:
    log = pkl.load(f)
```

```python
(gen_pretrain_train_loss,
    gen_pretrain_eval_loss,
    dis_pretrain_train_loss,
    dis_pretrain_train_acc,
    dis_pretrain_eval_loss,
    dis_pretrain_eval_acc,
    gen_adversarial_eval_loss,
    dis_adversarial_train_loss,
    dis_adversarial_train_acc,
    dis_adversarial_eval_loss,
    dis_adversarial_eval_acc) = log
```

# Training Settings

same as the default value in code

# Generator

```python
x1 = range(len(gen_pretrain_eval_loss))
plt.plot(x1, gen_pretrain_eval_loss)
plt.xlabel('Epochs')
plt.ylabel('Oracle NLL')
```

```python
x2 = range(len(gen_adversarial_eval_loss))
plt.plot(x2, gen_adversarial_eval_loss)
plt.xlabel('Epochs')
plt.ylabel('Oracle NLL')
```

```python
x3 = range(len(gen_pretrain_eval_loss) + len(gen_adversarial_eval_loss))
gen_loss = gen_pretrain_eval_loss + gen_adversarial_eval_loss
plt.plot(x3, gen_loss)
plt.xlabel('Epochs')
plt.ylabel('Oracle NLL')
plt.savefig('gen_loss.png',dpi=600)
```

# Discriminator

```python
x4 = range(len(dis_pretrain_eval_loss))
plt.plot(x4, dis_pretrain_eval_loss)
plt.xlabel('Steps (3 epoch per step)')
plt.ylabel('Oracle NLL')
```

```python
x5 = range(len(dis_adversarial_eval_loss))
plt.plot(x5, dis_adversarial_eval_loss)
plt.xlabel('Rounds (9 epoch per round)')
plt.ylabel('Oracle NLL')
```

```python
x6 = range(len(dis_pretrain_eval_loss) + len(dis_adversarial_eval_loss))
dis_loss = dis_pretrain_eval_loss + dis_adversarial_eval_loss
plt.plot(x6, dis_loss)
plt.xlabel('Steps/Rounds (3 epoch per step (x: 0-49), 9 epochs per round (x: 50-199))')
plt.ylabel('Oracle NLL')
plt.savefig('dis_loss.png',dpi=600)
```
